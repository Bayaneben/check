# Plot_correlation_map

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
```

The result of this function is a square matrix; each row represents a variable, and all the columns represent the same variables as rows; each cell in the grid represents the value of the correlation coefficient between two features(variables) that indicates how strong the relationship is and it consist of numbers between -1 and 1.

This correlation plot allows us to have a global view of the more or less strong relationship between several variables
It is a powerful tool to summarize a large dataset
